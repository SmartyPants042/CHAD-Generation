```python
from nltk import pos_tag
from nltk.corpus import stopwords
from string import punctuation
stop_words = set(stopwords.words('english'))
import json
```

```python
finals = []
```

```python
with open('cleaned_jokes.json') as file:
    data = json.load(file)
    for j,inp in enumerate(data):
        if j%10000==0:
            print(j)
        pun_in = inp['punch']
        set_in = inp['setup']
        for i in punctuation:
            pun_in = pun_in.replace(i,'')
            set_in = set_in.replace(i,'')
        set_in = set_in.lower()
        pun_in = pun_in.lower()
        set_in = set_in.split(' ')
        pun_in = pun_in.split(' ')
        senlen = len(set_in)+len(pun_in)
        try:
            puntags = pos_tag(pun_in)
            settags = pos_tag(set_in)
        except:
            continue
        set_temp = ''
        pun_temp = ''
        noun, verb, adj = 0,0, 0
        nouns, verbs, adjs = {},{},{}
        for i in settags:
            if i[0] in stop_words:
                set_temp+=i[0]+' '
            elif i[1][0]=='N':
                if i[0] in nouns.keys():
                    noun_n = nouns[i[0]]
                else:
                    noun+=1
                    noun_n = noun
                    nouns[i[0]] = noun
                set_temp+=f'NOUN_{noun_n} '
            elif i[1][0]=='V':
                if i[0] in verbs.keys():
                    verb_n = verbs[i[0]]
                else:
                    verb+=1
                    verb_n = verb
                    verbs[i[0]] = verb
                set_temp+=f'VERB_{verb_n} '
            elif i[1][0]=='J' or i[1][0]=='R':
                if i[0] in adjs.keys():
                    adj_n = adjs[i[0]]
                else:
                    adj+=1
                    adj_n = adj
                    adjs[i[0]] = adj
                set_temp+=f'ADJUNCT_{adj_n} '
            else:
                set_temp+=i[0]+' '
        for i in puntags:
            if i[0] in stop_words:
                pun_temp+=i[0]+' '
            elif i[1][0]=='N':
                if i[0] in nouns.keys():
                    noun_n = nouns[i[0]]
                else:
                    noun+=1
                    noun_n = noun
                    nouns[i[0]] = noun
                pun_temp+=f'NOUN_{noun_n} '
            elif i[1][0]=='V':
                if i[0] in verbs.keys():
                    verb_n = verbs[i[0]]
                else:
                    verb+=1
                    verb_n = verb
                    verbs[i[0]] = verb
                pun_temp+=f'VERB_{verb_n} '
            elif i[1][0]=='J' or i[1][0]=='R':
                if i[0] in adjs.keys():
                    adj_n = adjs[i[0]]
                else:
                    adj+=1
                    adj_n = adj
                    adjs[i[0]] = adj
                pun_temp+=f'ADJUNCT_{adj_n} '
            else:
                pun_temp+=i[0]+' '
        final = {}
        all_tags = noun+verb+adj
        if senlen//all_tags>3:
            print(set_temp)
            print(pun_temp)
            print()
        final['setup_temp'] = set_temp
        final['punch_temp'] = pun_temp
        final['noun'] = noun
        final['verb'] = verb
        final['adj'] = adj
        finals.append(final)
with open('templates', 'w') as fout:
    json.dump(finals, fout)     
        
```
